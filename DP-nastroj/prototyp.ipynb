```python
def compareLines(str1, str2): 
    if len(str1.split()) != len(str2.split()):
        return False
    for i in range(len(str1.split())):
        if str1.split()[i].isupper():
            continue
        if str1.split()[i-1] != str2.split()[i-1]:
            return False
    return True
```

```python
f = open("C://Users//Adam//Desktop//DP-v1.0//src//Main.elm", "r")
codeLines = f.readlines()
f.close
```

```python
checked = []
helper = 0
boilerplate = []
lines = []

for index, sentence in enumerate(codeLines):
    if len(codeLines[index].split()) < 2 or codeLines[index].split()[0] == "case" or index in checked:
        continue
    if "--" in codeLines[index].split()[0] or "_" in codeLines[index].split()[0] or "|" in codeLines[index].split()[0] or "=" in codeLines[index].split()[0] or codeLines[index].split()[0][0].isupper():
        continue
    reindex = index +1
    indexes = []
    while reindex != len(codeLines):
        if reindex in checked:
            reindex += 1
            continue
        if compareLines(codeLines[index], codeLines[reindex]):
            if len(indexes) == 0:
                boilerplate.append(codeLines[index])
                indexes.append(index+1)
            checked.append(reindex)
            indexes.append(reindex+1)
        reindex += 1
    if len(indexes) != 0:
        lines.append(indexes)
```

```python
for index, boiler in enumerate(boilerplate):
    print("boilerplate : " + str(boiler).replace("\n",""))
    print("na riadkoch : " + str(lines[index]))
    print("počet : " + str(len(lines[index])))
    print(" ")
```

```python
countLines = 0
countLinesWithoutSpaces = 0
countFunctions = 0

for subdir, dirs, files in os.walk("C://Users//Adam//Desktop//DP-v1.0//src//"):
    for filename in files:
        filepath = subdir + os.sep + filename
        
        if filepath.endswith(".elm"):
            countFunction = 0
            countCa = 0
            
            f = open(filepath, "r")
            codeLines = f.readlines()
            f.close
            
            countLines += len(codeLines)
            
            for line in codeLines:
                if line != "\n":
                    countLinesWithoutSpaces += 1
                    
                if line[0] == " ":
                    continue
                number = 0
                for word in line.split():
                    number += 1
                    if word == ":" and number == 2:
                        countFunctions += 1
                        countFunction += 1
                        
                if "import Components." in line:
                    for word in line.split():
                        if word.startswith("Components"):
                            countCa += 1
                            break
                    
            
            print(str(filepath).replace("C://Users//Adam//Desktop//DP-v1.0//src//",""))
            print("Počet riadkov kódu : " + str(len(codeLines)))
            print("Počet funkcií : " + str(countFunction))
            print("Počet riadkov kódu / Počet funkcií : " + str(len(codeLines) / countFunction))
            print("Ce : " + str(countCa))
            print(" ")

        else:
            continue
```
